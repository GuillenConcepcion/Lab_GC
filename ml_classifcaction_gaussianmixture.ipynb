# Modelos de mezclas gaussianas

sklearn.mixture es un paquete que permite aprender Modelos de mezcla gaussiana (diagonal, esférica, atada y de covarianza completa matrices compatibles), muestrearlas y estimarlas a partir de datos. Instalaciones para ayudar a determinar el número apropiado de También se proporcionan componentes.

# Librerias

```python
from numpy import unique
from numpy import where
from matplotlib import pyplot
from sklearn.datasets import make_classification
from sklearn.mixture import GaussianMixture
```

# Data

```python
# inicializar el conjunto de datos con el que trabajaremos
training_data, _ = make_classification(
    n_samples=10000,
    n_features=2,
    n_informative=2,
    n_redundant=0,
    n_clusters_per_class=1,
    random_state=4
)
```

# Model

```python
# definir el modelo
gaussian_model = GaussianMixture(n_components=2)
```

Traim

```python
# entrenar el modelo
gaussian_model.fit(training_data)
```

# Asignacion

```python
# asignar cada punto de datos a un grupo
gaussian_result = gaussian_model.predict(training_data)
```

# Clustering

```python
# obtener todos los grupos únicos
gaussian_clusters = unique(gaussian_result)
```

# Visualizacion

```python
# graficar los grupos de Mezcla Gaussiana
for gaussian_cluster in gaussian_clusters:
    # obtener todos los puntos de datos que caen en este grupo
    index = where(gaussian_result == gaussian_cluster)
    # hacer el gráfico
    pyplot.scatter(training_data[index, 0], training_data[index, 1])

# mostrar el gráfico de Mezcla Gaussiana
pyplot.show()
```
