# Definir función

```python
def  compute_projectile_distance ( initial_speed, angle ):
    """
    Calcula la distancia horizontal recorrida por un proyectil.

    Parámetros:
    initial_speed (float): Velocidad inicial del proyectil en m/s
    angle (float): Ángulo de lanzamiento en grados

    Devuelve:
    float: Distancia horizontal recorrida en metros
    """
    #Convertir ángulo de grados a radianes
    angle_rad = math.radians(angle)

    # Aceleración gravitacional (m/s²)
    g = 9.81

    # Calcula la distancia horizontal usando la fórmula de movimiento del proyectil:
    # distance = (initial_speed² * sen(2*angle)) / g
    distance = (initial_speed** 2 * math.sin( 2 * angle_rad)) / g
```
